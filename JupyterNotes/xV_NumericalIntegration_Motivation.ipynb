# Motivating Numerical Integration

Perhaps you remember integration being introduced to you in a way that was inherently numerical. The following table gives several methods which you were probably introduced to in order to approximate $\displaystyle{\int_{a}^{b}{f\left(x\right)dx}}$, where $x_0 = a$, $x_n = b$, and $x_0 < x_1 < x_2 < \cdots < x_n$.

<center>

Name | Expression for Approximating Integral
--- | ---
Left-Endpoint Method | $\displaystyle{\sum_{k = 0}^{n - 1}{f\left(x_i\right)\Delta x_i}}$
Right-Endpoint Method | $\displaystyle{\sum_{k = 1}^{n}{f\left(x_i\right)\Delta x_i}}$
Trapezoid Method | $\displaystyle{\sum_{k = 0}^{n - 1}{\frac{f\left(x_i\right) + f\left(x_{i+1}\right)}{2}\Delta x_i}}$

</center>



```python
#@title
x_vals = np.linspace(0, 10, 500)
y_vals = x_vals + np.sin(2*x_vals)

x_abscissas = np.linspace(0, 10, 10)
y_abscissas = x_abscissas + np.sin(2*x_abscissas)

plt.figure(figsize = (12, 4))
plt.plot(x_vals, y_vals, color = "purple")
plt.plot(x_abscissas, y_abscissas, color = "red")
plt.fill_between(x_abscissas, y_abscissas, color = "red", alpha = 0.2)
plt.scatter(x_abscissas, y_abscissas, color = "red", s = 25)
plt.vlines(x = x_abscissas, ymin = 0, ymax= y_abscissas, color = "red", ls = "--")
plt.grid()
plt.axvline(x = 0, color = "black")
plt.axhline(y = 0, color = "black")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trapezoidal Method for Approximating Integral of f(x) = x + sin(x) from x = 0 to x = 10")
plt.show()
```

In general, numerical integration is referred to as *quadrature*. We can approximate $\displaystyle{\int_{a}^{b}{f\left(x\right)dx} \approx \sum_{i = 0}^{n}{A_if\left(x_i\right)}}$, where the nodal abscissas ($x_i$) and the wieghts ($A_i$) depend on the particular rule being used for quadrature. **Note:** All rules of quadrature are derived from polynomial interpolation of the integrand, so these methods work best when $f\left(x\right)$ can be well-approximated by a polynomial.

In this section of our course, we'll methods of numerical integration falling under the umbrella of *Newton-Cotes formulae*. Newton-Cotes formulae require many function evaluations, and include common approaches such as those listed in the table above. 

The advantages to Newton-Cotes formulae are that they are based on local interpolation, so they only require piecewise fit to a polynomial. The disadvantages, however, are that they require many function evaluations, which can be expensive if $f\left(x\right)$ is difficult to compute.

Our textbook covers another set of techniqes, termed *Gaussian Quadrature*. In *Gaussian Quadrature*, the locations of the abscissas are chosen to increase accuracy of the approximation of the integral. These techniques require fewer function evaluations to achieve a desired level of accuracy. These techniques can also be used to handle integrable singularities, such as $\displaystyle{\int_{0}^{1}{\frac{g\left(x\right)}{\sqrt{1 - x^2}} dx}}$ for well-behaved functions $g\left(x\right)$.
