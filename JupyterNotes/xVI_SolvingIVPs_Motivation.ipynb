# Motivating Numerical Solutions to Initial Value Problems

The goal of this section of our course is to explore the construction of numerical solutions to

$$\vec{y}' = \vec{F}\left(x, \vec{y}\right)~~\text{where}~~\vec{y}\left(a\right) = \vec{\alpha}$$

```python
#@title
x, y = np.meshgrid(np.linspace(-5, 5, 25),
				np.linspace(-5, 5, 25))

def dy_dx(x, y):
  return np.sin(x - y)

# Directional vectors
v = dy_dx(x, y)
magnitude = np.sqrt(1 + v**2)
u = np.ones_like(v)

#y0 = [-2.75, 0.25]
#x_sol = np.linspace(-5, 5, 250)
#y_sol = solve_ivp(dy_dx, y0, x_sol).y

#print(y_sol)

# Plotting Slope Field with QUIVER
plt.figure(figsize = (12, 4))
plt.quiver(x, y, u/magnitude, v/magnitude, color='purple')
#plt.plot(x_sol, y_sol, color = "green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Slope Field for dy/dx = sin(x - y)")

# Setting x, y boundary limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Show plot with grid
plt.grid()
plt.show()
```

## Differential Equations of Different Orders

As a reminder, the *order* of a differential equation is the highest-order derivative of $y$ present in the equation. For example, a *first-order differential equation* takes the form $\displaystyle{y' = f\left(x, y\right)}$. A general *$n^{th}$-order differential equation* takes the form $\displaystyle{y^{\left(n\right)} = f\left(x, y, y', y'', \cdots, y^{\left(n-1\right)}\right)}$

## Notes and Notation

The following note and notation will be useful throughout our discussions on solving IVPs.

**Note:** Any $n^{th}$-order differential equation, $\displaystyle{y^{\left(n\right)} = f\left(x, y, y', \dots, y^{\left(n-1\right)}\right)}$, can be decomposed into a series of first-order differential equations.

$$\left\{\begin{array}{lcl} y_0 & = & y\\ y_1 & = & y'\\ y_2 & = & y''\\ & \vdots & \\ y_{n - 1} & = & y^{\left(n-1\right)}
\end{array}\right. \implies \left\{\begin{array}{lcl} y_0' & = & y_1\\ y_1' & = & y_2\\ y_2' & = & y_3\\ & \vdots & \\ y_{n - 1}' & = & y_{n}\\ y_n' & = & f\left(x, y_0, y_1, \cdots, y_{n-1}\right)
\end{array}\right.$$

with initial conditions

$$\left\{\begin{array}{lcl} y_0\left(a\right) & = & \alpha_0\\ y_1\left(a\right) & = & \alpha_1\\ y_2\left(a\right) & = & \alpha_2\\ & \vdots & \\ y_{n - 1}\left(a\right) & = & \alpha_{n-1}
\end{array}\right.$$

**Notation:** Using vector notation for our differential equations will allow us to manipulate sets of first-order differential equations in a concise form. For example, we can write $\vec{y}' = \vec{F}\left(x, \vec{y}\right)$ with $\vec{y}\left(a\right) = \vec{\alpha}$ where

$$\vec{F}\left(x, \vec{y}\right) = \left[\begin{array}{c}y_1\\ y_2\\ \vdots\\ y_n\\ f\left(x, \vec{y}\right)\end{array}\right]$$

## Initial Conditions

The solution to any $k^{th}$-order differential equation contains $k$ arbitrary constants. We can obtain a unique solution if we have a set of $k$ auxiliary constraints of the form $\vec{y}\left(a\right) = \vec{\alpha}$ such that $y_i\left(a\right) = \alpha_i$ means $y^{\left(i\right)}\left(a\right) = \alpha_i$ for $i = 0, 1, \cdots, k-1$.

As long as $x = a$ is the consistent value for $x$ involved in all of these "initial conditions", then the differential equation is referred to as an *initial value problem* (IVP). If the initial conditions occur at different values of $x$, then the problem is referred to as a *boundary value problem*.

For example, $\left\{\begin{array}{lcl} y\left(0\right) & = & 0\\ y'\left(0\right) & = &  1\\ y'' & = & -y\end{array}\right.$ is an IVP, while $\left\{\begin{array}{lcl} y\left(\pi\right) & = & 0\\ y'\left(0\right) & = &  1\\ y'' & = & -y\end{array}\right.$ is a *boundary value problem*. We'll only consider IVPs here. Boundary value problems are more difficult to solve and may be left as an option for final course projects.
