## Project Requirements

Below are the requirements for our MAT370 course reflection.

  + In a few paragraphs, describe what numerical methods / numerical analysis means to you. You might include answers to the following in your discussion.
  
    + Do you expect to apply numerical techniques in your remaining coursework?
    + Do you wish you had access to these numerical techniques in any of your prior coursework? Which ones and why?
    + Were there points in the semester where you felt empowered by the numerical approaches we covered?
    + Were there points in the semester where you felt less comfortable (perhaps even frustrated) with numerical approaches than the algebraic techniques you learned in previous coursework?
    + What is your level of comfort with Python coming out of this course? Do you expect to continue using Python in the future? If so, what do you hope to get better at?

  + Revisit two problems that you found interesting from earlier in the semester. These problems may have been homework problems or problems from our unit problem sets.
  
    + Write up full, detailed solutions to these problems. Your solutions should include an introduction to the problem and, in the event that equations/functions were provided to you, ~~a discussion of where those mathematical objects arose from~~ a discussion of the components of the equations/functions (you do not need necessarily need to derive them). You should also document your solution, describing what each step (in particular any code) is accomplishing.
  
  + Your project should include references and any inline citations, if you do any research beyond our course notes or textbook. The method of citation is yours to choose -- you may go full APA or MLA, or may choose [IEEE citation format](https://ieee-dataport.org/sites/default/files/analysis/27/IEEE%20Citation%20Guidelines.pdf) which is common in mathematics and the sciences.
