```python
#@title
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
x_vals = np.random.uniform(low = 0, high = 10, size = 3)
y_vals = (x_vals - 7)**2
```

# Interpolation and Curve-Fitting

Having spent time considering how to solve linear systems, we'll move on to our next topic: *Interpolation and Curve Fitting*. Don't forget about solutions to linear systems though, we'll need some of that work here as well!

## Interpolation

Consider that we have $n$ observed data points. Recall that we can fit a polynomial of degree $n-1$ that runs through all of these points exactly. In fact, that polynomial is unique!

```python
#@title
x_new = np.linspace(0, 10, num = 250)
y_plot = (x_new - 7)**2

plt.figure(figsize = (12, 4))

plt.subplot(1, 3, 1)
plt.scatter(x_vals, y_vals)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed Data")

plt.subplot(1, 3, 3)
plt.plot(x_new, y_plot, color = "purple")
plt.scatter(x_vals, y_vals)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed Data and Interpolated Curve")

plt.show()
```

A polynomial interpolation of $n$ observed data points is a unique polynomial of degree at most $n-1$ such that $p\left(x^*\right) = y^*$ for all $\left(x^*, y^*\right)$ in the set of observed data. If you've looked at the code in this notebook, you've noticed that I cheated by using a polynomial to construct my observed data and then just plotted that generating polynomial through the points. Can we find/construct an interpolating function (polynomial or otherwise) that will fit a set of observed data when we don't know the genrating function ahead of time?

```python
#@title
x_vals = np.random.uniform(0, 10, 5)
y_vals = np.random.uniform(0, 10, 5)

plt.figure(figsize = (4,4))
plt.scatter(x_vals, y_vals)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed Data")
plt.show()
```

## Curve Fitting

Unlike interpolation, *curve-fitting* seeks to model inherently noisy data. That is, we aren't seeking to find a curve that fits through each of the observed data points -- instead, we are trying to construct a line/curve of best fit which captures the general trend between independent variables and a corresponding dependant variable. Many of you will have *heard* of linear regression before, and some of you may have even used it.

```python
#@title
x_vals = np.random.uniform(low = 0, high = 10, size = 20)
y_vals = 2*x_vals + 8 + np.random.normal(loc = 0, scale = 4, size = len(x_vals))

x_new = np.linspace(0, 10, num = 100)
y_new = 2*x_new + 8

plt.figure(figsize = (12, 4))
plt.subplot(1, 3, 1)
plt.scatter(x_vals, y_vals)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed Data")

plt.subplot(1, 3, 3)
plt.plot(x_new, y_new, color = "purple")
plt.scatter(x_vals, y_vals)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed Data and Regression Line")

plt.show()
```

Again, if you've looked at the code in the notebook, you've noticed that I cheated by using a known function to generate my observed data. I simply plotted the general function (without the noise) that I used to generate my $\left(x, y\right)$ pairs. Can we still discover the data generating function if we do not know it ahead of time, and only know of the observed data?

### Looking Ahead

Over the next several class meetings, our focus will be on numerical approaches to interpolation and to curve fitting. In the interest of clarity, the plot below shows the difference between an *interpolating function* (red) and curve fitting (green).

```python
#@title
x_vals = np.random.uniform(0, 10, 4)
y_vals = (x_vals-4)*(x_vals-7)**2

x_new = np.linspace(0, 10, 100)
y_interp = (x_new - 4)*(x_new - 7)**2
lin_fit = np.polyfit(x_vals, y_vals, deg = 1)
lin_fit_fn = np.poly1d(lin_fit)
y_fit = lin_fit_fn(x_new)

plt.figure(figsize = (12, 4))
plt.scatter(x_vals, y_vals)
plt.plot(x_new, y_interp, color = "red")
plt.plot(x_new, y_fit, color = "green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolating Function (Red) and Curve Fit (Green)")
plt.ylim((-100, 50))
plt.show()
```
